```python
#extracting each product's url from first 3 pages

product_urls = []
for i in urls:
    
    driver.get(i)
    product_url = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in product_url:
        product_urls.append(i.get_attribute('href'))
    
        
    
```

```python
len(product_urls)
```

```python
brand_names = []
product_names = []
ratings = []
no_of_ratings = []
price = []
return_exchanges = []
expected_delivery_dates = []
availability = []
other_details = []
```

```python
count = 1
for i in product_urls:
    driver.get(i)
    print(count)
    count = count +1
    #brand name 
    try:
        try:
            product_name_obj  = driver.find_element_by_xpath("//span[@id='productTitle']")
            product_names.append(product_name_obj.text)
        except NoSuchElementException as e:
            brand_names.append('-')

        #product name 

        try:
            brand_obj = driver.find_element_by_xpath("//a[@id='bylineInfo']")
            brand_names.append(brand_obj.text)
        except NoSuchElementException as e:
            product_names.append('-')



        # rating  feature
        try:
            rating_obj = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
            ratings.append(rating_obj.text)
        except NoSuchElementException as e:
             ratings.append('-')

        #number of ratings  

        try:
            no_of_ratings_obj =driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
            no_of_ratings.append(no_of_ratings_obj.text)
        except NoSuchElementException as e:
            no_of_ratings.append('-')

        #Price    

        try:
            price_obj =driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
            price.append(price_obj.text)
        except NoSuchElementException as e:
            price.append('-')

        #Expected Delivery Date

        try:
            expected_delivery_obj = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b")
            expected_delivery_dates.append(expected_delivery_obj.text)

        except NoSuchElementException as e:
            expected_delivery_dates.append('-')

        #Availability (In stock/ not)
        try:
            available_obj = driver.find_element_by_xpath("//div[@id='availability']/span")
            availability.append(available_obj.text)
        except NoSuchElementException as e:
            availability.append('-')
            
        # exchange-return policy
        try:
        exchange_obj = driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']/span/div/a")
        exchange_return.append(exchange_obj.text)
    except NoSuchElementException:
        exchange_return.append('-')

        # OTher Details Pay on Delivery, 7 Days Replacement, Amazon Delivered, Warranty Period


        try:
            other_details_obj = driver.find_elements_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
            li = []
            for j in other_details_obj:

                li.append(j.text)

            other_details.append(', '.join(li))
        except NoSuchElementException as e:
            other_details.append('-')
    except :
        print('ERROR')

```

    Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”.

    and CSV.


```python
url = 'https://www.flipkart.com'
driver.get(url)
```

```python
product_urls = []
for i in product_link:
    product_urls.append(i.get_attribute('href'))
```

```python
len(product_urls)
```

```python
product_names = []
details = []
colors = []
prices = []

for i in product_urls:
    driver.get(i)
    try:
        #product name/ brand
        product_name = driver.find_element_by_xpath("//h1[@class='yhB1nd']/span")
        product_names.append(product_name.text)
    except NoSuchElementException:
        product_names.append('-')
    # Color
    try:
        color = driver.find_element_by_xpath("//div[@class=\"_3k-BhJ\"][1]/table/tbody/tr[4]/td[2]/ul/li")
        colors.append(color.text)
    except NoSuchElementException:
        colors.append('-')
    try:
        specs = driver.find_elements_by_xpath("//li[@class='_21Ahn-']")
        count = 0
        li = []
        for j in specs:
            li.append(j.text)
            
        details.append(li)
        #print(li)
            

    except NoSuchElementException:
        print('-')
    # Price
    try:
        price = driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
        prices.append(price.text)
    except NoSuchElementException:
        prices.append('-')
        
        
    
   
        

```

```python
len(details),len(prices),len(colors), len(product_names)
```

```python
# Breaking Details into many infors of RAM ROM DISPLAY, CAMERa
ram = []
rom = []
display_size = []
display_res = []
camera_front = []
camera_rear = []
battery = []
processor = []
for i in range(24):
    #Ram AND ROM
    li = (details[i][0]).split('|')
    ram.append(li[0])
    rom.append(li[1])
    
    #DIsplay Size and Resolution
    li_display = (details[i][1]).split()
    d_size = li_display[:4]
    d_res = li_display[4:]
    display_size.append(' '.join(d_size))
    display_res.append(' '.join(d_res))
    
    #Camera front and Rear
    camera_info = (details[i][2]).split('|')
    if(len(camera_info)==2):
        camera_front.append(camera_info[0])
        camera_rear.append(camera_info[1])
    if(len(camera_info)==1):
        camera_front.append(camera_info[0])
        camera_rear.append('-')
    #Battery
    battery.append(details[i][3])
    #PRocessor
    if(len(details[i])>=5):
        processor.append(details[i][4])
    else:
        processor.append('-')
brand = []        
for i in product_names:
    brand.append(i.split()[0])
    
        
    
```

```python
dict = {
    'Product Name': product_names,'Brand':brand, 'Color': colors, 'Price':prices,
    'RAM': ram, 'ROM': rom,'Display Size':display_size, 'Display Resolution':display_res,
    'Battery': battery,'Front Camera': camera_front,'Camera REAR': camera_rear,
    'Processor': processor, 'Product URL': product_urls
}
table = pd.DataFrame(dict)
table
```
